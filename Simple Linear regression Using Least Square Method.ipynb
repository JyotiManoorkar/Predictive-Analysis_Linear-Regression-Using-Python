```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=(20.0,10.0)

```

```python
#Reding Data
data=pd.read_csv('C:\JyotiLearnings\GitDemo\Predictive-Analysis_Linear-Regression-Using-Python\headbrain.csv')
```

```python
print(data.shape)
```

```python
data.head()
```

```python
#Collect X and Y
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
#Mean X and Y

mean_X=np.mean(X)
mean_Y=np.mean(Y)
# total number of values
m=len(X)
# using the formula to calculate b1 and b2
numer=0
denom=0

for i in range(m):
    numer+=(X[i]-mean_X)*(Y[i]-mean_Y)
    denom+=(X[i]-mean_X)**2
b1=numer/denom
b0=mean_Y-(b1*mean_X)
# print coefficients
print(b1)
print(b0)
# this is our linear model
```

```python
#lets plot model -plotting values and regression line

maxi_x=np.max(X)+100
mini_x=np.min(X)-100
#Calculating line values of x and y
x=np.linspace(mini_x,maxi_x,1000)
y=b0+b1*x
#plotting a line
plt.plot(x,y,color='Blue',label='Regression Line')
#Plotting scatter points
plt.scatter(X,Y,c='#ef5423',label='Scatter Plot'
           )
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.legend()
plt.show()
```

```python
#R Sqaure value
ss_t=0
ss_r=0
for i in range(m):
    y_pred=b0+b1*X[i]
    ss_t+=(Y[i]-mean_Y)**2
    ss_r+=(Y[i]-y_pred)**2
    r2=1-(ss_r/ss_t)
print(r2)   # here we have implemented Simple Linear Regression model using least square method
```

```python
# implement model using machine learning library scikit
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# cannot use Rank 1 matrix in scikit learn
X=X.reshape(m,1)

# creating model
reg=LinearRegression()
#fitting a data
reg=reg.fit(X,Y)
#Y prediction
Y_pred=reg.predict(X)
#calculate  R2 score
r2_score=reg.score(X,Y)
print(r2_score)
#calculate additional RMSE
mse=mean_squared_error(Y,Y_pred)
rmse=np.sqrt(mse)
print(rmse)
```
